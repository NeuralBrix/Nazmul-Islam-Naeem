```python
import pandas as pd
```

```python
data = {

    "name": ["Alice", "Bob", "Charlie", "David", "Eva", "Frank", "Grace", "Hannah", "Ian", "Jack"],
    "age": [23, 45, 31, 35, 29, 40, 28, 33, 38, 410],
    "math": [85, 78, 92, 88, 76, 95, 89, 84, 91, 87],
    "science": [90, 82, 88, 91, 79, 94, 85, 87, 93, 80],
    "english": [78, 85, 80, 82, 88, 90, 84, 79, 81, 83]
        
}
```

```python
df =pd.DataFrame(data)
```

```python
print("math marks :",df.math.tolist())
print("math mean :",df.math.mean())
print("math median :",df.math.median())     
```

```python
print("Math Mean:", df["math"].mean())
```

```python
Q1 = df["math"].quantile(0.25)
Q3 = df["math"].quantile(0.75)
IQR = Q3 - Q1

lower_fence = Q1 - 1.5 * IQR
upper_fence = Q3 + 1.5 * IQR

print("Math - Q1:", Q1, "Q3:", Q3, "IQR:", IQR)
print("Lower Fence:", lower_fence, "Upper Fence:", upper_fence)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.boxplot(x=df["math"])
plt.title("Math Marks Boxplot")
plt.show()
```

```python
sns.boxplot(x=df["science"], color="skyblue")
plt.title("Science Marks Boxplot", fontsize=14, fontweight='bold')
plt.xlabel("Science Marks", fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```

```python
colors = ["skyblue", "lightgreen", "orange"]
sns.boxplot(data=df[["math","science","english"]], palette=colors)
plt.title("Colored Marks Boxplot", fontsize=30, fontweight='bold',color='purple')
plt.show()
```
